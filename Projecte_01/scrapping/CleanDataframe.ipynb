```python
fusion_columnas = {
    "Memoria RAM": ["RAM", "Capacidad de la memoria RAM", "Capacidad de la memoria RAM instalada", "Memoria RAM"],
    "Marca": ["Marca", "Fabricante"],
    "Tipo de pantalla": ["Tipo de pantalla", "Tecnologia de pantalla"]
}

for col_final, cols_a_unir in fusion_columnas.items():
    cols_existentes = [col for col in cols_a_unir if col in df.columns]
    
    if cols_existentes:
        df[col_final] = df[cols_existentes].bfill(axis=1).iloc[:, 0]
        columnas_a_borrar = [col for col in cols_existentes if col != col_final]
        df.drop(columns=columnas_a_borrar, inplace=True)
```
